```python
! pip -q install scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro, norm, binom, pearsonr, ttest_ind

# ---- sanity: required columns (rename here if your file differs) ----
req = [
    "anxiety_level","self_esteem","mental_health_history","depression",
    "headache","blood_pressure","sleep_quality","breathing_problem",
    "noise_level","living_conditions","safety","basic_needs",
    "academic_performance","study_load","teacher_student_relationship",
    "future_career_concerns","social_support","peer_pressure",
    "extracurricular_activities","bullying","stress_level"
]
missing = [c for c in req if c not in df.columns]
if missing:
    raise ValueError(f"Missing columns in df: {missing}")

# ---------------------------
# 1) NORMALITY CHECK (Academic Performance)
# ---------------------------
x = df["academic_performance"].astype(float)

# Shapiro-Wilk test
W, p_shapiro = shapiro(x.sample(min(len(x), 500), random_state=42))  # shapiro limit-friendly
print(f"[Normality] Shapiro-Wilk W={W:.3f}, p-value={p_shapiro:.4f}")
print("  -> p>0.05 suggests data may be approximately normal.")

# Histogram + fitted normal curve
mu, sigma = x.mean(), x.std(ddof=1)
plt.figure(figsize=(6,4))
count, bins, _ = plt.hist(x, bins=10, density=True, alpha=0.6)
xs = np.linspace(min(x), max(x), 200)
pdf = norm.pdf(xs, mu, sigma)
plt.plot(xs, pdf, linewidth=2)
plt.title("Academic Performance: Histogram with Normal PDF")
plt.xlabel("academic_performance"); plt.ylabel("density")
plt.show()

# ---------------------------
# 2) BINOMIAL PROBABILITY
#    P(at least 7 'high stress' in a group of 10)
#    Define 'high stress' from your label format.
# ---------------------------
def is_high_stress(series):
    if series.dtype == 'O':  # strings like "Distress"/"Eustress"/"No Stress"
        return series.str.lower().str.contains("distress", na=False)
    # numeric labels: assume 2 = Distress (highest), 1 = Eustress, 0 = No stress
    return series.astype(int) >= 2

high_mask = is_high_stress(df["stress_level"])
p_hat = high_mask.mean()
n, k = 10, 7
prob_at_least_k = binom.sf(k-1, n, p_hat)  # 1 - cdf(k-1)
print(f"[Binomial] Empirical p(high stress)={p_hat:.3f}")
print(f"  P(X ≥ {k}) for X~Binomial(n={n}, p={p_hat:.3f}) = {prob_at_least_k:.4f}")

# ---------------------------
# 3) CONDITIONAL PROBABILITY
#    P(depression ≥ 4 | academic_performance ≤ 2)
# ---------------------------
low_acad = df["academic_performance"] <= 2
dep_high = df["depression"] >= 4
if low_acad.sum() > 0:
    cond_prob = (dep_high & low_acad).sum() / low_acad.sum()
    print(f"[Conditional] P(depression≥4 | academic_performance≤2) = {cond_prob:.3f}")
else:
    print("[Conditional] No students with academic_performance ≤ 2 in the data.")

# ---------------------------
# 4) CORRELATION: anxiety vs academic performance (+ scatter & trend line)
# ---------------------------
x_anx = df["anxiety_level"].astype(float)
y_acad = df["academic_performance"].astype(float)
r, p_corr = pearsonr(x_anx, y_acad)
print(f"[Correlation] Pearson r(anxiety, academic_performance) = {r:.3f}, p={p_corr:.4g}")
print("  -> Negative r means higher anxiety relates to lower performance.")

plt.figure(figsize=(6,4))
plt.scatter(x_anx, y_acad, s=18)
# regression line
m, b = np.polyfit(x_anx, y_acad, 1)
xs = np.linspace(x_anx.min(), x_anx.max(), 200)
plt.plot(xs, m*xs + b, linewidth=2)
plt.xlabel("anxiety_level"); plt.ylabel("academic_performance")
plt.title("Anxiety vs Academic Performance")
plt.show()

# ---------------------------
# 5) HYPOTHESIS TEST:
#    Mean academic_performance: Distress vs Eustress
#    (works for string or numeric labels)
# ---------------------------
def group_mask(level_name_or_code):
    s = df["stress_level"]
    if s.dtype == 'O':  # string labels
        return s.str.lower().str.contains(level_name_or_code, na=False)
    # numeric labels
    mapping = {"distress": 2, "eustress": 1}
    code = mapping[level_name_or_code]
    return s.astype(int) == code

g_distress = df.loc[group_mask("distress"), "academic_performance"].astype(float)
g_eustress = df.loc[group_mask("eustress"), "academic_performance"].astype(float)

if len(g_distress) > 1 and len(g_eustress) > 1:
    tstat, p_tt = ttest_ind(g_distress, g_eustress, equal_var=False)
    print(f"[T-test] Distress vs Eustress (academic_performance): t={tstat:.3f}, p={p_tt:.4g}")
    print(f"  Means -> distress={g_distress.mean():.2f}, eustress={g_eustress.mean():.2f}")
else:
    print("[T-test] Not enough samples in one of the groups to run the test.")

# ---------------------------
# 6) Quick factor importance view by stress type (bar of means)
# ---------------------------
factor_cols = [
    "anxiety_level","self_esteem","depression","sleep_quality","study_load",
    "social_support","peer_pressure","bullying","academic_performance"
]
present = [c for c in factor_cols if c in df.columns]
means_by_stress = df.groupby("stress_level")[present].mean()

print("\n[Means by stress_level] (first few rows)")
print(means_by_stress.head())

for col in present:
    plt.figure(figsize=(6,4))
    means_by_stress[col].plot(kind="bar")
    plt.title(f"Average {col} by stress_level")
    plt.xlabel("stress_level"); plt.ylabel(f"avg {col}")
    plt.xticks(rotation=0)
    plt.show()

```
